```python
#filling missing values 1. with zeroes 
df_filled=data.fillna(0)
df_filled.isnull().sum()
```

```python
#filling missing values with mean value of particular column

#CUDA, Metal, OpenCL,Vulkan this columns has missing values and all column's data type is float
data['CUDA_filledna'] = data['CUDA'].fillna(data['CUDA'].mean())
data['Metal_filledna']=data['Metal'].fillna(data['Metal'].mean())
data['OpenCL_filledna']=data['OpenCL'].fillna(data['OpenCL'].mean())
data['Vulkan_filledna']=data['Vulkan'].fillna(data['Vulkan'].mean())
```

```python
data.head(5)
```

```python
#rename the columns
data=data.rename(columns={'Manufacturer':'Manufac'})
data.head(4)
```

```python
## change data types 
data['CUDA_filledna']=data['CUDA_filledna'].astype(int)
#if column has missing values it will give error
```

```python
data.dtypes
```

```python
#applying a function to a column value 
#let say metal usage is decreased or half
data['model_metal']=data['Metal_filledna'].apply(lambda x:x/2)
data.head(4)
```

```python
#Data aggregating and grouping
grp_mean=data.groupby('Device')['Metal_filledna'].mean()
grp_mean.head(2)
```

```python
# group by based on two columns
grp_data= data.groupby(['Device']).sum()
```

```python
#aggregate functions - max(), min() value
max(data['Metal_filledna'])
min(data['CUDA_filledna'])
```

create two sample data frame and lets merge them

```python
df1=pd.DataFrame({'key':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'Value2':[4,5,6]})
```

```python
# Merging and Joining data frame 
#merge data frame on common column 
pd.merge(df1,df2,on="key",how="right") # on="" this parameter decide on which column they get merged
# how="inner" this will show which type of join
# use left or outer in place of right 
```
